```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load and Read in CSV Squirrel Data
squirrel_df = pd.read_csv('Resources/nyc_squirrels_2.csv')  
squirrel_df = squirrel_df.loc[:, ~squirrel_df.columns.str.contains('^Unnamed')]
squirrel_df
```

```python
#duplicateRows = squirrel_df[squirrel_df.duplicated()]
ids = squirrel_df["unique_squirrel_id"]
duplicates = squirrel_df[ids.isin(ids[ids.duplicated()])].sort_values("unique_squirrel_id")
duplicates
# 5 Rows identifed as duplicates. Drop to arrive at accurate counts. 
```

```python
# Remove duplicate squirrel observations
squirrel_df = squirrel_df.drop_duplicates(subset=['unique_squirrel_id'])
squirrel_df
```

```python
# Check for duplicates again
ids = squirrel_df["unique_squirrel_id"]
duplicates = squirrel_df[ids.isin(ids[ids.duplicated()])].sort_values("unique_squirrel_id")
duplicates
```

```python
# Calculate total of each color type
gray_sq = squirrel_df[squirrel_df["primary_fur_color"] == "Gray"]
gray_sq_count = len(squirrel_df[squirrel_df["primary_fur_color"] == "Gray"])
cinnamon_sq_count = len(squirrel_df[squirrel_df["primary_fur_color"] == "Cinnamon"])
black_sq_count = len(squirrel_df[squirrel_df["primary_fur_color"] == "Black"])
print(gray_sq_count)
print(cinnamon_sq_count)
print(black_sq_count)
```

```python
# Calculate percentage of each squirrel type
gray_percent = len(squirrel_df[squirrel_df["primary_fur_color"] == "Gray"])/(squirrel_df['primary_fur_color'].count()) * 100
cinnamon_percent = len(squirrel_df[squirrel_df["primary_fur_color"] == "Cinnamon"])/(squirrel_df['primary_fur_color'].count()) * 100
black_percent = len(squirrel_df[squirrel_df["primary_fur_color"] == "Black"])/(squirrel_df['primary_fur_color'].count()) * 100
print(gray_percent)
print(cinnamon_percent)
print(black_percent)
```
